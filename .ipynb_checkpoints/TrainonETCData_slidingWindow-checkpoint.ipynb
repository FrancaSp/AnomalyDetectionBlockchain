```python
from Prediction import Prediction
from DataPreparation import DataPreparation
import pandas as pd
from tensorflow.keras.models import load_model
from matplotlib import pyplot as plt
import os
from matplotlib.patches import Rectangle
import numpy as np
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
```

```python
df1 = pd.read_csv('/Users/francaspeth/Google Drive/Masterarbeit/Share/ETC_block_2018_withuncles.csv')
df2 = pd.read_csv('/Users/francaspeth/Google Drive/Masterarbeit/Share/ETC_block_2019_withuncles.csv')
df3 = pd.read_csv('/Users/francaspeth/Google Drive/Masterarbeit/Share/ETC_block_2020_withuncles.csv')

valid = pd.concat([df2,df3], ignore_index= True)
train = valid.loc[(valid['number'] > 7261690) & (valid['number'] < 10904146)]
valid = pd.concat([df1, df2], ignore_index = True)
valid = valid.loc[(valid['number'] < 7245623)]
del(df1,df2,df3)
```

```python
timestep = 10
PrepTrain = DataPreparation(train, timestep)
PrepTrain.df.head()
PrepTrain.df.sort_values('number')
PrepTrain.change_idx()
PrepValid = DataPreparation(valid, timestep)
```

```python
cols = ['num_tx', 'avg_tx_value', 'avg_nonce', 'size', 'difficulty',
       'total_difficulty', 'gas_limit', 'gas_used','num_emptyblocks', 'miner', 'extra_data',
       'num_uncles','total_num_uncles'] # 'num_emptyblocks',
PrepTrain.df = PrepTrain.df.loc[:,cols]
PrepTrain.drop_first_n_cols()
PrepTrain.change_data_types(cols)
trainp = PrepTrain.scale()
trainp = np.reshape(trainp, (int(len(PrepTrain.df) / timestep), timestep, len(PrepTrain.df.columns)))

PrepValid.df = PrepValid.df.loc[:,cols]
PrepValid.drop_first_n_cols()
PrepValid.change_data_types(cols)
validp = PrepValid.scale()
validp = np.reshape(validp, (int(len(PrepValid.df) / timestep), timestep, len(PrepValid.df.columns)))
```

```python
model = load_model('/Users/francaspeth/Google Drive/Masterarbeit/Share/AutoencoderModelle/LSTM/TrainingonETC/tumblingWindow/AutoencoderETC10Timestep16Apr2021110817.h5'
                  , compile = False)
```

```python
pred = model.predict(trainp)
testMAE = np.mean(np.abs(pred - trainp), axis=1)
plt.figure(figsize=(10, 10))
plt.hist(testMAE, bins=30, label = cols)
plt.legend(prop={'size': 10})
plt.show()
```

```python
pred = model.predict(validp)
testMAE = np.mean(np.abs(pred - validp), axis=1)
plt.figure(figsize=(10, 10))
plt.hist(testMAE, bins=30, label=cols)
plt.legend(prop={'size': 10})
plt.show()
```
